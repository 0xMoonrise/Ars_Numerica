```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def graph(function, min_value=-5.0, max_value=5.0,
          value=10, name='', inter=None, asin=None):
    x = np.linspace(min_value, max_value, value)
    y = function(x)

    fig, ax = plt.subplots()
    ax.plot(x, y, label=name)

    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')

    ax.set_xlabel("x")
    ax.set_ylabel("y")

    ax.yaxis.set_label_coords(-0.08, 0.5)

    if inter is not None:
        for point in inter:
            x_inter, y_inter = point
            ax.plot(x_inter, y_inter, 'ro', label=f'Punto de intersección: ({x_inter:.2f}, {y_inter:.2f})', zorder=3)
            #ax.annotate(f'({x_inter:.2f}, {y_inter:.2f})', (x_inter, y_inter),
            #            textcoords="offset points", xytext=(5,5), ha='center')
    if asin is not None:
        ax.axhline(y=asin, color='black', linestyle='--',
                   label='Asíntota horizontal', zorder=2)
    ax.legend(loc='upper left')
    plt.show()

    num_rows = value
    step = len(x) // num_rows
    data = {'x': x[::step], 'y': y[::step]}
    df = pd.DataFrame(data)

    print("Tabla de Datos:")
    print(df)
```
